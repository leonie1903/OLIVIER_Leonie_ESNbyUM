```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk

# stopwords et le tokenizer de NLTK
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# fichier CSV
data = pd.read_csv("train.csv")

# Vérifier existance des colonnes 
if 'Context' not in data.columns or 'Response' not in data.columns:
    raise ValueError("Assurez-vous que votre fichier CSV contient les colonnes 'Context' et 'Response'.")

# fonction pour traiter la colonne et générer un Counter des mots
def process_column(column_data):
    text_data = " ".join(column_data.astype(str).tolist())
    tokens = word_tokenize(text_data.lower())  # Convertir en minuscules et tokeniser
    tokens = [word for word in tokens if word.isalnum()]  # Garder uniquement les mots alphanumériques
    stop_words = set(stopwords.words("english"))  # Charger les stopwords en anglais
    filtered_tokens = [word for word in tokens if word not in stop_words]  # Supprimer les stopwords
    return Counter(filtered_tokens)

# Traiter les colonnes "Context" et "Response" séparément
context_counts = process_column(data['Context'])
response_counts = process_column(data['Response'])

# Combiner les deux colonnes et traiter
combined_counts = context_counts + response_counts

# Afficher les mots les plus fréquents pour chaque cas
def display_most_common(word_counts, title, top_n=10):
    most_common_words = word_counts.most_common(top_n)
    print(f"\nLes {top_n} mots les plus fréquents pour {title} :")
    for word, freq in most_common_words:
        print(f"{word}: {freq}")
    return most_common_words

display_most_common(context_counts, "Context")
display_most_common(response_counts, "Response")
display_most_common(combined_counts, "l'ensemble (Context + Response)")

# Générer des nuages de mots
def generate_wordcloud(word_counts, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white', 
                          max_words=200, colormap='viridis').generate_from_frequencies(word_counts)
    plt.figure(figsize=(15, 7.5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

# Nuage de mots pour chaque colonne et l'ensemble
generate_wordcloud(context_counts, "Nuage de mots pour Context")
generate_wordcloud(response_counts, "Nuage de mots pour Response")
generate_wordcloud(combined_counts, "Nuage de mots pour l'ensemble (Context + Response)")
```

```python
from textblob import TextBlob

# fonction pour calculer le score de polarité
def analyze_sentiment(text):
    if isinstance(text, str):
        return TextBlob(text).sentiment.polarity
    return None

# Appliquer l'analyse de sentiment sur les colonnes "Context" et "Response"
data['Context_Polarity'] = data['Context'].apply(analyze_sentiment)
data['Response_Polarity'] = data['Response'].apply(analyze_sentiment)

# Visualiser la distribution des scores de polarité avec des histogrammes
def plot_sentiment_distribution(polarity_data, title):
    plt.figure(figsize=(10, 6))
    plt.hist(polarity_data, bins=20, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel("Score de polarité")
    plt.ylabel("Fréquence")
    plt.grid(axis='y', alpha=0.75)
    plt.show()

# Distribution des scores de sentiment pour chaque cas
plot_sentiment_distribution(data['Context_Polarity'].dropna(), "Distribution des scores de polarité pour Context")
plot_sentiment_distribution(data['Response_Polarity'].dropna(), "Distribution des scores de polarité pour Response")

# Comparaison question reponses 
print("\nExemple 1 :")
print(f"Context : {data['Context'][0]}")
print(f"Response : {data['Response'][0]}")
print(f"Polarité du contexte : {data['Context_Polarity'][0]}")
print(f"Polarité de la réponse : {data['Response_Polarity'][0]}")


print("\nExemple 2 :")
print(f"Context : {data['Context'][1]}")
print(f"Response : {data['Response'][1]}")
print(f"Polarité du contexte : {data['Context_Polarity'][1]}")
print(f"Polarité de la réponse : {data['Response_Polarity'][1]}")

```
